```python
players = pd.read_csv('players.csv')
countries = pd.read_csv('countries.csv')
events = pd.read_csv('events.csv')
```

```python
p75 = players[players['birthdate'].str.split('-').str[0] == '1975']
p75_gy = p75[['name']].groupby(p75['country_id']).size()
pl75_df = pd.DataFrame({'country_id' : list(p75_gy.index.values), 'num_count' : list(p75_gy)})
pd.merge(countries, pl75_df, on='country_id', how='left').fillna(0)[['name', 'num_count']]
```
