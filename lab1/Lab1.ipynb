```python
from sklearn import svm
from sklearn.datasets import load_iris
import math

# Load the iris dataset
iris = load_iris()

n_samples_iris, n_features_iris = iris.data.shape

num_ele = int(-1.0 * math.ceil(n_samples_iris * 0.2))
#print num_ele

# Split the data into training/testing sets
iris_X_train = iris.data[:num_ele]
iris_X_test = iris.data[num_ele:]

# Split the targets into training/testing sets
iris_y_train = iris.target[:num_ele]
iris_y_test = iris.target[num_ele:]

# Create linear SVM 
clf_linear = svm.SVC(kernel='linear')

# Fit linear SVM
clf_linear.fit(iris_X_train, iris_y_train)

# Predict class for testing data
pred = clf_linear.predict(iris_X_test)

clf_non_linear = svm.SVC(kernel='rbf')

# Fit the svm
clf_non_linear.fit(iris_X_train, iris_y_train)

# Calculate prediction error for testing data
nlpred = clf_non_linear.predict(iris_X_test)

# Calculate prediction error for testing data
errors = 0.0
for i in range(len(pred)):
    if pred[i] != iris_y_test[i]:
        errors += 1.0
print "Linear: Test data errors = ",errors, "out of", len(pred), "predictions"

# Calculate prediction error for all points
total_pred_linearSVM = clf_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred_linearSVM)):
    if total_pred_linearSVM[i] != iris.target[i]:
        errors += 1.0
print "Linear: Total errors = ",errors, "out of", len(total_pred_linearSVM), "predictions"

errors = 0.0
for i in range(len(nlpred)):
    if nlpred[i] != iris_y_test[i]:
        errors += 1.0
print "Non-Linear: Test data errors = ",errors, "out of", len(nlpred), "predictions"

# Calculate prediction error for all points
total_pred_non_linearSVM = clf_non_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred_non_linearSVM)):
    if total_pred_non_linearSVM[i] != iris.target[i]:
        errors += 1.0
print "Non-Linear: Total errors = ",errors, "out of", len(total_pred_non_linearSVM), "predictions"
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import math

# Load the iris dataset
iris = load_iris()

n_samples_iris, n_features_iris = iris.data.shape

num_ele = int(-1.0 * math.ceil(n_samples_iris * 0.2))

# Split the data into training/testing sets
iris_X_train = iris.data[:num_ele]
iris_X_test = iris.data[num_ele:]

# Split the targets into training/testing sets
iris_y_train = iris.target[:num_ele]
iris_y_test = iris.target[num_ele:]

# Create decision tree 
dt_clf = DecisionTreeClassifier(random_state=0)

# Build decision tree
dt_clf.fit(iris_X_train, iris_y_train)

# Predict class for testing data
pred = dt_clf.predict(iris_X_test)

# Calculate prediction error for testing data
errors = 0.0
for i in range(len(pred)):
    if pred[i] != iris_y_test[i]:
        errors += 1.0
print "Decision Tree: Test data errors = ",errors, "out of", len(pred), "predictions"

# Calculate prediction error for all points
total_pred_dt = dt_clf.predict(iris.data)
errors = 0.0
for i in range(len(total_pred_dt)):
    if total_pred_dt[i] != iris.target[i]:
        errors += 1.0
print "Decision Tree: Total errors = ",errors, "out of", len(total_pred_dt), "predictions"

```
